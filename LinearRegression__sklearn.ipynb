```python
#Importing library
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
boston=datasets.load_boston()
```

```python
#Type of boston data
type(boston)
```

```python
print(boston.keys())
```

```python
#Shape of boston data
boston.data.shape
```

```python
#Printing feature_names
boston.feature_names
```

```python
#Creating data frame 
x=boston.data
y=boston.target
data=pd.DataFrame(x,columns=boston.feature_names)

#Describtion of data
data.describe()
```

```python
#Plotting scatter matrix, c imput for target variiable
pd.scatter_matrix(data,c=y)
plt.show()
```

```python
#Applying regression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=12)

from sklearn.linear_model import LinearRegression

# Appling Regression
reg = LinearRegression()

#Fitting regression model
reg.fit(x_train,y_train)

#Predicting 
reg.predict(x_test)

#R square 
reg.score(x_test,y_test)



```

```python
#Lets try cross validation
from sklearn.model_selection import cross_val_score
cv=cross_val_score(reg,x,y,cv=3)
print (cv)
```

```python
#Mean of cross validation
np.mean(cv)
```
