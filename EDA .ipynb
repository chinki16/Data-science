```python
#Reading the data
file='C:/Users/chink/Google Drive/Computational Statistics/3rd Quater/Statistical learning with R/Homework/insurance.csv'
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv(file)
data.head(2)
```

```python
#Boxplot of variables
data.plot(kind='box',subplots=True)
plt.show()
```

```python
data.plot(y='expenses',kind='hist',bins=30,normed=True)
plt.xlabel("Expenses in USD")
plt.show()
```

```python
#Summary statistics 
data.describe()
```

```python
#Standard deviation of all data values 
data.std()
```

```python
#Check for quantiles 
q=[0.25,0.75]
data.quantile(q)
```
