```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=6)
#Fitting data
knn.fit(iris['data'],iris['target'])
#Predicting data
predict=knn.predict(iris['data'])
predict
```

```python
#KNN algorithm
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris['data'],iris['target'],test_size=0.3,random_state=21)
knn=KNeighborsClassifier(n_neighbors=8)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
knn.score(x_test,y_test)
```

```python
#Creating arrays
neighbors=np.arange(1,9)
print(type(neighbors))

#Defining empty train accuracy array
train_accuracy=np.empty(len(neighbors))
#Defining empty test accuracy array
test_accuracy=np.empty(len(neighbors))

#Different value of k 
# emumerate return i index and k value at 
for i,k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    #Fitting model
    knn.fit(x_train,y_train)
    #Calculating accuracy
    train_accuracy[i]=knn.score(x_train,y_train)
    test_accuracy[i]=knn.score(x_test,y_test)
    
#Plotting test accuracy
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.legend()
#Plotting test accuracy
plt.plot(neighbors, train_accuracy, label = 'Testing Accuracy')
plt.show()   
```
